```python
import pandas as pd
import numpy as np

# aggregated_results is produced by gencsv.py, which read all SUMMARY_* files and parse them into one CSV
# aggregated_results_new.csv contains data for newly added target jerryscript, php, z3
data_path = "aggregated_results_undirected.csv"
data = pd.read_csv(data_path)
data.head()
```

```python
# check the unique values in the column
for column in data.columns:
    if column in ["file", "time"]:
        continue
    print(f"{column}: {data[column].unique()}")
```

```python
# rows with final_result == 'B'
data_bug = data[data["final_result"] == "B"]
# keep only rows with different value in column 'commit'
data_bug = data_bug.drop_duplicates(subset=["iid"])
data_bug
```

```python
data_reach = data[data["final_result"] == "R"]
data_reach = data_reach.drop_duplicates(subset=["commit"])
# only display columns: file, iid, commit, final_result
data_reach = data_reach[["file", "iid", "commit", "final_result"]]
data_reach
```

```python
data_behave = data[data["final_result"] == "D"]
data_behave = data_behave.drop_duplicates(subset=["commit"])
data_behave = data_behave[["file", "iid", "commit", "final_result"]]
data_behave
```
